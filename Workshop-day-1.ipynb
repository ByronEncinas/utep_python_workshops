### How to work with files

image.png

```python
# file object

# we want to write on numbers.txt

file = open("/content/drive/MyDrive/Python Workshop SPS/numbers.txt", "w")

# we will create a list with nunmbers

num = [5,4,3,2,1,0]

# we are going to write all of this values.

# we need to get all of the values on the list, one each time.

for elem in num:

  # but all data that goes into a file has to be a string, so...

  new_var = str(elem)

  print(new_var)

  file.writelines(new_var)

  file.flush()

file.close()
```
